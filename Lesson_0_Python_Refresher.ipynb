## Lists

Lists are one of the most basic data structures in Python. They are used to store collections of related values, such as a list of names, a list of numbers, or a list of objects. Lists are defined using square brackets [ ], with each element in the list separated by a comma ,.

For example, here is a partial list of cities in the United States:



```python
# define a list of cities
cities = ["New York", "Los Angeles", "Chicago", "Houston", "Philadelphia", "Cambridge", "San Francisco", "2"]
print(cities)
```

You can access individual elements in a list using their index, which starts at 0 for the first element in the list. For example, to access the second element in the cities list above (which is "Los Angeles"), you would use the following code:

```python
your_city = cities[0]
my_city = cities[6]

print( "and your favorite city is ", my_city, " Great!!!")
```

### Syntax of a list

- Square brackets []: Lists are defined using square brackets to enclose the contents of the list.

- Commas ,: The contents of the list consist of a series of values, separated by commas.

- Indexing: Each value in the list is assigned an index based on its position in the list, starting at 0 for the first value and increasing by 1 for each subsequent value. You can access individual values in the list using their index.

- Length: You can determine the number of values in a list using the len() function.

### Manipulating Lists

Methods: Lists have a number of built-in methods that allow you to manipulate and interact with the list. Some common methods include append(), insert(), remove(), and sort().

Iteration: You can iterate over the contents of a list using a loop, such as a for loop, to perform some operation on each value in the list.

Here are some simple code examples for the append(), insert(), remove(), and sort() methods of lists in Python:

```python
# Define a list of numbers
my_list = [3, 1, 4, 1, 5, 9, 2, 6]

# Append a value to the end of the list
my_list.append(8)
print(my_list)  # Output: [3, 1, 4, 1, 5, 9, 2, 6, 8]

# Insert a value at a specific position in the list
my_list.insert(2, 7)
print(my_list)  # Output: [3, 1, 7, 4, 1, 5, 9, 2, 6, 8]

# Remove a value from the list
my_list.remove(5)
print(my_list)  # Output: [3, 1, 7, 4, 1, 9, 2, 6, 8]
# Sort the list in ascending order
my_list.sort()
print(my_list)  # Output: [1, 1, 2, 3, 4, 6, 7, 8, 9]
```

### Results

In the example above, we start with a list of numbers my_list = [3, 1, 4, 1, 5, 9, 2, 6].

- We use the append() method to add the value 8 to the end of the list.

- We use the insert() method to insert the value 7 at position 2 in the list.

- We use the remove() method to remove the value 5 from the list.

- Finally, we use the sort() method to sort the list in ascending order.

Each of these methods modifies the list in some way, and we print out the result of each method call to see the updated contents of the list

## Zero Indexing

Zero indexing is a convention used in many programming languages, including Python, where the first element of an array, list, or tuple is assigned an index of 0, rather than 1. This means that the first element in a sequence is located at index 0, the second element is located at index 1, and so on.

For example, if we define a list of strings like this:   **my_list = ["apple", "banana", "kiwi", "orange"]**

Then the string "apple" is located at index 0, "banana" is located at index 1, "kiwi" is located at index 2, and "orange" is located at index 3. The total length of the list is 4, but the valid indices for accessing elements in the list are 0, 1, 2, and 3.

Zero indexing can take some getting used to, especially if you are coming from a background in mathematics where indices typically start at 1. However, it is an important convention to understand when working with arrays, lists, and tuples in Python.



## Tuples

Tuples are similar to lists in that they allow you to store collections of related values. However, tuples are immutable, meaning that once they are created, their contents cannot be changed. Tuples are defined using parentheses ( ), with each element in the tuple separated by a comma ,

For example, here is a tuple representing a person's name, age, and occupation:

```python
# Tuples

person = ("John Smith", 35, "Engineer")

print(person[0])
```

You can access individual elements in a tuple using their index, just like with lists

### Tuble Syntax Explained

Here's a breakdown of the syntax of a tuple in Python:

- Parentheses (): Tuples are defined using parentheses to enclose the contents of the tuple.

- Commas ,: The contents of the tuple consist of a series of values, separated by commas.

- Indexing: Each value in the tuple is assigned an index based on its position in the tuple, starting at 0 for the first value and increasing by 1 for each subsequent value. You can access individual values in the tuple using their index.

- Immutability: Tuples are immutable, which means that once a tuple is defined, its contents cannot be modified. This is in contrast to lists, which are mutable and can be modified after they are defined.

- Packing and Unpacking: You can assign multiple values to a tuple at once using packing, and you can assign the values of a tuple to multiple variables using unpacking.

## Dictionaries
Dictionaries are used to store collections of key-value pairs, where each key maps to a corresponding value. Dictionaries are defined using curly braces {}, with each key-value pair separated by a colon : and each pair separated by a comma ,.

For example, here is a dictionary mapping the names of fruits to their colors:

```python
# Dictionaries

fruit_colors = {"apple": "red", "banana": "yellow", "kiwi": "green", "orange": "orange"}

print(fruit_colors)  # the print statement recognizes the data type and prints accordingly
```

You can access the value associated with a key in a dictionary using square brackets []. For example, to get the color of an apple from the fruit_colors dictionary above, you would use the following code:

```python
apple_color = fruit_colors["kiwi"]

print(apple_color)
```

In the context of the dictionary example above, a key-value pair refers to a single entry in the dictionary.

In this example, each entry consists of two parts: a "key" and a "value". The "key" is a unique identifier that is used to look up the corresponding "value". In the fruit_colors dictionary, for example, the keys are the names of fruits (e.g. "apple", "banana", "kiwi", etc.), and the values are the colors of those fruits (e.g. "red", "yellow", "green", etc.).

The syntax of a dictionary in Python consists of several key components:

- Curly braces {}: Dictionaries are defined using curly braces to enclose the contents of the dictionary.

- Key-value pairs: The contents of a dictionary are made up of key-value pairs, where each key-value pair maps a unique "key" to a corresponding "value".

- Colons :: The key-value pair itself consists of two parts: the "key" and the "value", separated by a colon. The key is always on the left side of the colon, and the value is always on the right.
